```python
covid_df_2020=covid_df[covid_df["year"]==2020]
covid_df_2021=covid_df[covid_df["year"]==2021]
covid_df_2022=covid_df[covid_df["year"]==2022]
covid_df_2020.shape,covid_df_2021.shape,covid_df_2022.shape
```

```python
lst1=[]
lst2=[]
lst_death=[]
for i in covid_df_2020["month"].unique():
    lst1.append(covid_df_2020[covid_df_2020["month"]==i]["cases_total"].mean())
    lst_death.append(covid_df_2020[covid_df_2020["month"]==i]["deaths_total"].mean())
    #print(df.shape)
    lst2.append(i)
df_2020=pd.DataFrame()
df_2020["Month"]=lst2
df_2020["Avg Cases"]=lst1
df_2020["Avg Deaths"]=lst_death
df_2020=df_2020.sort_values(by="Month")
df_2020
```

```python
import seaborn
import matplotlib.pyplot as plt

# declaring data
data = [mean_0_17, mean_18_29, mean_30_39, mean_40_49, mean_50_59,mean_60_69, mean_70_79 ,mean_80]
keys = ['Age 0-17', 'Age 18-29', 'Age 30-39', 'Age 40-49', 'Age 50-59','Age 60-69','Age 70-79','Age 80+']
fig = plt.figure(figsize = (20, 10))
# declaring exploding pie
explode = [0, 0, 0, 0, 0,0.1,0,0]
# define Seaborn color palette to use
palette_color = seaborn.color_palette('dark')
  
# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')
#plt.legend()
plt.title("Agewise distribution of Hospitalization ")
# displaying chart
plt.show()
```
