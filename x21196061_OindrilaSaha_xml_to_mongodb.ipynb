```python
import xml.etree.ElementTree as etr
```

```python
def file_parser(path):
    try:
        flightparse=etr.parse(path)
        flight=flightparse.getroot()
        return flight
    except FileNotFoundError:
        print("Given File not available".format(path))
    except IOError:
        print("Unable to read from given file".format(path))
    finally:
        print("File parsing complete")
        
path=r'C:\Users\Oindrila\OneDrive\Desktop\1st Trimester\Data Analytics Programming\project\hiv_aids_dataset.xml'
HIV_AIDS_OBJ=file_parser(path)
```

```python
final_list=[]
for obj in HIV_AIDS_OBJ.findall('row/'):
#     year=obj.find('year').text
#     print(year)
    year=obj.find('year').text
    neighborhood=obj.find('neighborhood').text
    sex=obj.find('sex').text
    race=obj.find('race').text
    hiv_diagnoses_num=obj.find('hiv_diagnoses_num').text
    hiv_diagnoses_num_per_100k=obj.find('hiv_diagnoses_num_per_100k').text
    hiv_aids_concurrent_num=obj.find('hiv_aids_concurrent_num').text
    concurrent_hiv_aids_among_all=obj.find('concurrent_hiv_aids_among_all').text
    aids_diagnoses_num=obj.find('aids_diagnoses_num').text
    aids_diagnoses_num_per_100k=obj.find('aids_diagnoses_num_per_100k').text
    print(year,neighborhood,sex,race,hiv_diagnoses_num,hiv_diagnoses_num_per_100k,hiv_aids_concurrent_num,concurrent_hiv_aids_among_all,aids_diagnoses_num,aids_diagnoses_num_per_100k)
    final_list.append({"year":year,
                     "neighborhood" : neighborhood,
                     "sex":sex,
                     "race":race,
                     "hiv_diagnoses_num":hiv_diagnoses_num,
                     "hiv_diagnoses_num_per_100k": hiv_diagnoses_num_per_100k,
                     "hiv_aids_concurrent_num": hiv_aids_concurrent_num,
                     "concurrent_hiv_aids_among_all": concurrent_hiv_aids_among_all,
                     "aids_diagnoses_num": aids_diagnoses_num,
                     "aids_diagnoses_num_per_100k": aids_diagnoses_num_per_100k,
                      })
    print("------------------------------------------------")
```
