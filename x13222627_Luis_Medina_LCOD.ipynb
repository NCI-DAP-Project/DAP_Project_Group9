```python
def file_parser(path):
    try:
        flightparse=etr.parse(path)
        flight=flightparse.getroot()
        return flight
    except FileNotFoundError:
        print("Given File not available".format(path))
    except IOError:
        print("Unable to read from given file".format(path))
    finally:
        print("File parsing complete")
        
path=r'C:\Users\teach\OneDrive\Escritorio\NCI\Database & Analytics Programming\CA1\Leading_cause_of_death.xml'
lcod_df=file_parser(path)
```
