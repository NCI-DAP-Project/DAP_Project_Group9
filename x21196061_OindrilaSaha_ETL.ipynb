```python
import pymongo
import pandas as pd
from pymongo import MongoClient
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

```python
def data_extraction(url,dbname,collectname):
    client = MongoClient()
    client = MongoClient(url)
    db = client[dbname]
    test = db[collectname]
    df = pd.DataFrame(list(test.find()))
    return df
```

```python
hiv_df=data_extraction("mongodb://localhost:27017/",'DAP','HIV_AIDS')
```
