```python
import pymongo
import pandas as pd
from pymongo import MongoClient
import psycopg2
import numpy as np
#from sqlalchemy import create_engine
```

```python
def data_extraction(url,dbname,collectname):
    client = MongoClient()
    client = MongoClient(url)
    db = client[dbname]
    test = db[collectname]
    df = pd.DataFrame(list(test.find()))
    return df
```
